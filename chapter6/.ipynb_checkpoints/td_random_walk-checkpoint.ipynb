```python
import numpy as np
```

```python
v = [0, 1/6, 2/6, 3/6, 4/6, 5/6, 1]  # v(s), the true values that we're estimating

# params / settings
alpha = 0.1
gamma = 0.9
episodes = [0, 3, 10, 50, 100]

# policy: random left/right with p=0.5
```

```python
historical_Vs = []
```

```python
for num_episodes in episodes:
    # V(s) where s in [A, B, C, D, E]; what we're estimating
    V = [0, 0.5, 0.5, 0.5, 0.5, 0.50, 0]  # initialization
    
    for episode in range(num_episodes):
        state = 3  # start at C
        while state > 0 and state < 6:
            # get action from policy (randomly choose left/right with p=0.5)
            action = -1 if np.random.uniform() < 0.5 else 1

            # "take action" and get reward
            new_state = state + action
            reward = 1 if new_state == 6 else 0

            # update value function
            V[state] = V[state] + alpha * (reward + gamma*(V[new_state] - V[state]))

            # next time step
            state = new_state

        # terminal state
        #print(f'\nReached terminal state for episode: {episode}, state: {state}')
    
    print(f'V after all episodes: {V}\n')
    historical_Vs.append(V)
```

```python
import matplotlib.pyplot as plt
```

```python
bar = [0, 1, 2, 3]
bar[1:-1]
```

```python
fig, ax = plt.subplots()

for V in historical_Vs:
    ax.plot(V[1:-1])
    
# Plot true
ax.plot(v[1:-1])
    
ax.set_xlabel('States [A, B, C, D, E]')
ax.set_ylabel('Estimated V(s)')
ax.set_title('Estimations for different # of episodes')

legend_list = [f'E={e}' for e in episodes]
legend_list.append('True')

plt.legend(legend_list)
plt.xticks(range(len(v[1:-1])))

plt.show()
```

TD(0) learning gets better as you sample from more episodes, up to a certain point.
